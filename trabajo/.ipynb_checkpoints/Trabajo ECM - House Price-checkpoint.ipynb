```python
import pandas as pd
import numpy as np

df= pd.read_csv ('House_price.csv', sep=',')

print(df.columns) ##COMPROBAMOS QUE SE HAYA LEÍDO 

df = df.drop(columns=['Address'])

## funciones para redondear los datos
df['Number of Rooms'] = np.ceil(df['Number of Rooms'])
df['Number of Bedrooms'] = np.ceil(df['Number of Bedrooms'])

##comprobamos que se ha redondeado de forma correcta
print(df)
```

```python
df.describe()
```

```python
import statsmodels.api as sm
# Modelo Regresión:  modeldata=stock.values 
X=df.values[:,0:5] # LArge B/P, LArge ROE, Large S/P, Large Return Rate Qt, Large Market Value, Small System Risk
y=df.values[:,5] # Anual Return
print("\nVARIABLE X\n")
print(X)
print("\nVARIABLE Y\n")
print(y)
```

```python
results = sm.OLS(y, sm.add_constant(X)).fit()

results.summary()
```
